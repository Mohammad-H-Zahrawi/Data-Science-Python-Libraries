## 3. Find distance between two points in 3D-space

Distance-Between-Two-Points-in-Three-Dimensions.png

```python
def length ():
  x=input('enter the coordinates of first point seperated by spaces')
  y=input('enter the coordinates of first point seperated by spaces')
  x1 , y1 , z1 = x.split(' ')
  x2 , y2 , z2 = y.split(' ')
  x1 = int(x1)
  y1 = int(y1)
  z1 = int(z1)
  x2 = int(x2)
  y2 = int(y2)
  z2 = int(z2)

  length = sqrt((x2-x1)**2+(y2-y1)**2+(z2-z1)**2)
  return length
```

```python
def thermal():
  print('1. linear expansion')
  print('2. surface expansion')
  print('3. volume expansion')
  x = int(input('what do you want to calculate?'))


  if x==1:
    L0 = float(input('initial length'))
    alpha = float(input('linear expansion coefficient? '))
    T1 = float(input('initial temperature? '))
    T2 = float(input('final temperature? '))
    dL = L0 * alpha * (T2-T1)
    return dL

  elif x==2:
    A0 = float(input('initial area'))
    beta = float(input('surface expansion coefficient? '))
    T1 = float(input('initial temperature? '))
    T2 = float(input('final temperature? '))
    dA = L0 * beta * (T2-T1)
    return dA

  elif x==3:
    V0 = float(input('initial volume'))
    gamma = float(input('volumetric expansion coefficient? '))
    T1 = float(input('initial temperature? '))
    T2 = float(input('final temperature? '))
    dV = V0 * gamma * (T2-T1)
    return dV

  else:
    print('please enter 1 or 2 or 3')

```

## 4. The Ideal gas law formula

The Ideal gas law formula states that

**PV = nRT**

Where P is the pressure (atm); V is the volume (L); T is temperature (K), n is the number of moles; R is a constant (R = 0.082)

**Example**:

What variable do you want to calculate?

[1] Pressure

[2] Volume

[3] Temperature

[4] n of mols


User:  1

Volume? X

Temperature? Y

N of moles? Z


-> the Pressure is W atms

```python
R = 0.082
print('1. pressure')
print('2. Volume')
print('3. Temperature')
print('4. Moles')
var = int(input('which variable do you want to calculate? '))


if var != 1 and var !=2 and var !=3 and var !=4:
  print('invalid option, choose 1, 2, 3, 4')
else:
  if var==1:
    #pressure
    n= float(input('Number of moles: '))
    T= float(input('Temperature (K): '))
    V = float(input('Volume (L)'))
    P = (n*R*T)/V
    print('the pressure is', P, 'atm')
  
  elif var==2:
    #volume
    n= float(input('Number of moles: '))
    T= float(input('Temperature (K): '))
    P = float(input('Pressure (atm): '))
    V = (n * R * T)/P
    print('the volume is', V, 'L')

  elif var==3:
    #temperature
    P = float(input('Pressure (atm): '))
    n= float(input('Number of moles: '))
    V = float(input('Volume (L'))
    T = (P*V)/(n*R)
    print('the temperature is', T, 'K')

  elif var==4:
    #moles
    P = float(input('Pressure (atm): '))
    V = float(input('Volume (L'))
    T= float(input('Temperature (K): '))
    n = (P*V)/(R*T)
    print('the number of moles is', n, 'K')




     
```

**Temperature records**


Create a Python program that asks the user how many temperatures he/she wants to record. Next the program must ask what type of unit the user will use (ºC or ºF). The program must add these temperatures in a list. 
From the first list, it should also create a second list with the temperatures stored in ºK 

**Use the following formulas:**

```python
def converator():
  print('1. cel_to_kel')
  print('2. kel_to_cel')
  print('3. fah_to_kel')
  print('4. kel_to_fah')

  x = input('enter conversion type ')

  if x=='1':
    cel = float(input('enter the temperature in Celsius '))
    kel = cel + 273.15
    return kel

  if x=='2':
    kel = float(input('enter the temperature in Kelvin '))
    cel = kel - 273.15
    return cel

  if x=='3':
    fah = float(input('enter the temperature in Fahrenheit '))
    kel = (fah - 32) * (5/9) + 273.15
    return kel

  if x=='4':
    kel = float(input('enter the temperature in kelvin '))
    kel = (kel - 273.15)*(9/5)+32
    return kel
```

```python
question_status = []

count=0
score=0
while count < len(trials):
  if correct[count]==trials[count]:
    score += 1
    question_status.append('correct')
  else:
    question_status.append('wrong')
  count += 1
```

```python
for question in question_status:
  print(f'Question status: {question}')

print(f'Overall score {score}/5')
```
